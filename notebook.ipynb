```python
import pandas as pd
import numpy as np
import os
```

```python
# Danh s√°ch topics
TOPICS = [
    'X√£ h·ªôi', 'Th·∫ø gi·ªõi', 'Kinh t·∫ø', 'ƒê·ªùi s·ªëng', 'S·ª©c kho·∫ª',
    'Gi√°o d·ª•c', 'Th·ªÉ thao', 'Gi·∫£i tr√≠', 'Du l·ªãch',
    'Ph√°p lu·∫≠t', 'Khoa h·ªçc - C√¥ng ngh·ªá', 'Xe'
]

def read_all_csvs_to_tuples():
    all_tuples = []
    
    for topic in TOPICS:
        # ƒê∆∞·ªùng d·∫´n file
        file_path = f'datasets/{topic.lower()}.csv'
        
        # Ki·ªÉm tra file c√≥ t·ªìn t·∫°i kh√¥ng
        if not os.path.exists(file_path):
            print(f"‚ö†Ô∏è  File kh√¥ng t·ªìn t·∫°i: {file_path}")
            continue
        
        try:
            # ƒê·ªçc CSV (kh√¥ng coi d√≤ng ƒë·∫ßu l√† header)
            df = pd.read_csv(file_path, encoding='utf-8', header=None)
            
            # L·∫•y c·ªôt 0 v√† c·ªôt 1, t·∫°o tuple
            tuples = list(zip(df[0], df[1]))
            
            # Th√™m v√†o list ch√≠nh
            all_tuples.extend(tuples)
            
            print(f"‚úÖ {topic}: {len(tuples)} tuples")
            
        except Exception as e:
            print(f"‚ùå L·ªói ƒë·ªçc {topic}: {e}")
    
    # Chuy·ªÉn th√†nh numpy array
    result_array = np.array(all_tuples)
    
    print(f"\nüìä T·ªîNG K·∫æT:")
    print(f"  - T·ªïng s·ªë tuples: {len(result_array)}")
    print(f"  - Shape c·ªßa array: {result_array.shape}")
    return result_array

# Ch·∫°y function
data_array = read_all_csvs_to_tuples()
```
